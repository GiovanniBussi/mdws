# Analyze Zenodo statistics

## Import packages

```python
import matplotlib.pyplot as plt
import pandas as pd
import yaml

# Update font size for graphics
plt.rcParams.update({'font.size': 20})
```

```python
zenodo = pd.read_csv("zenodo_stats.csv", sep=",")
zenodo
```

## Statistics

```python
zenodo.describe()
```

```python
zenodo[["size_per_year_in_TB", "datasets_per_year", "files_per_year"]].sum()
```

## Number of files and datasets per year

```python
fig, ax1 = plt.subplots(figsize=(10,7))
color = "tab:red"
ax1.set_xlabel("Year")
ax1.set_ylabel("Number of files", color=color)
ax1.plot(zenodo["year"], zenodo["files_per_year"], color=color, marker="o")
ax1.tick_params(axis="y", labelcolor=color)

ax2 = ax1.twinx()

color = "tab:orange"
ax2.set_ylabel("Number of datasets", color=color)
ax2.plot(zenodo["year"], zenodo["datasets_per_year"], color=color, marker="s")
ax2.tick_params(axis="y", labelcolor=color)

fig.tight_layout()  # otherwise the right y-label is slightly clipped
```

## Number of files and size per year

```python
fig, ax1 = plt.subplots(figsize=(10,7))
color = "tab:red"
ax1.set_xlabel("Year")
ax1.set_ylabel("Number of files", color=color)
ax1.plot(zenodo["year"], zenodo["files_per_year"], color=color, marker="o")
ax1.tick_params(axis="y", labelcolor=color)

ax2 = ax1.twinx()

color = "tab:blue"
ax2.set_ylabel("Size (TB)", color=color)
ax2.plot(zenodo["year"], zenodo["size_per_year_in_TB"], color=color, marker="s")
ax2.tick_params(axis="y", labelcolor=color)

fig.tight_layout()  # otherwise the right y-label is slightly clipped
```
